## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
Decision: Which pages should an SEO editor review and rewrite metadata for first?
Action: The editor opens the top-ranked pages and rewrites the title/meta description, then monitors whether CTR improves.
Cost of a wrong call: Flagging a fine page wastes an editor's review time (Mistake A) — annoying but cheap. Missing a page that's genuinely leaking clicks (Mistake B) means real opportunity goes unnoticed indefinitely — I consider this the more expensive mistake, so I'd rather my ranking lean toward flagging more candidates (recall) than being overly conservative (pure precision).

```

```python
import pandas as pd
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

visible = df[df["impressions_90d"] >= 100]  # enough volume to trust the CTR
by_tier = visible.groupby("position_tier")["ctr"].agg(["mean", "count"]).sort_values("mean", ascending=False)
print(by_tier)

gap = by_tier["mean"].max() - by_tier["mean"].min()
print(f"\nCTR gap between best and worst position tier: {gap:.3f} percentage points")

merged = visible.merge(by_tier["mean"].rename("tier_mean_ctr"), on="position_tier")
underperformers = (merged["tier_mean_ctr"] > merged["ctr"]).sum()
print(f"Pages sitting below their own tier's average CTR: {underperformers} of {len(merged)}")
```
